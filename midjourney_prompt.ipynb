## Define Prompt Generator

```python
import ipywidgets as widgets
from IPython.display import display
from google.colab import drive
from datetime import datetime
import csv

drive.mount("/content/drive", force_remount=True)
csv_path = '/content/drive/My Drive/prompt_archive.csv'

def read_and_print_csv():
    with open(csv_path, mode='r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            print(*row)

def mj_prompts_generator():
    # Function to handle the "Generate" button click event
    def generat_prompt_from_input(wid:None=None):
        # Function to handle the "Generate" button click event
        start_prompt = "/imagine prompt:"
        selected_prompts = []
        selected_ban_prompts = []
        selected_properties = []
        selected_to_choose = []

        # Process image prompts and weights
        for widget, key in zip(image_widgets, image_weight_widgets.keys()):
            image_value = widget.value
            image_weight_value = image_weight_widgets[key].value
            if image_value:
                selected_prompts.append(f'{image_value}::{int(image_weight_value)}')

        # Process ban prompts
        for ban_widget in ban_widgets:
            ban_value = ban_widget.value
            if ban_value:
                selected_ban_prompts.append(f"--no {ban_value}")

        # Process properties
        for key, widget in select_widgets_dict.items():
            select_value = widget.value
            if select_value:
                if key in properties_dict.keys():
                    selected_properties.append(f"--{key} {select_value}")
                else:
                    selected_to_choose.append(f"{key} {select_value}")


        # Join the selected prompts, properties, and ban prompts
        all_prompts = ", ".join(selected_prompts)
        all_ban_prompts = " ".join(selected_ban_prompts)
        all_properties = " ".join(selected_properties)
        all_selected = ", ".join(selected_to_choose)

        # Construct and print the final prompt
        final_prompt = f"{start_prompt} {all_prompts},\n{all_selected} {all_properties} {all_ban_prompts}"
        today_date = datetime.today().strftime('%Y-%m-%d')

        data = [[final_prompt, today_date]]

        with open(csv_path, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(data)

        print(final_prompt)

    properties_widgets_dict = {name: widgets.Dropdown(options=options, description=name + ":") for name, options in properties_dict.items()}

    choose_widgets_dict = {name: widgets.Dropdown(options=options, description=name + ":") for name, options in choose_dict.items()}

    select_widgets_dict = {**choose_widgets_dict, **properties_widgets_dict}

    # Creating text area widgets for entering and ban images, one for each input
    num_image_widgets = len(select_widgets_dict) - 3
    image_widgets = [widgets.Textarea(value='', description=f'image{num}:', layout=widgets.Layout(height='25px')) for num in range(num_image_widgets)]
    ban_widgets = [widgets.Textarea(value='', description=f'ban{num}:', layout=widgets.Layout(height='25px')) for num in range(3)]
    image_n_ban_widgets = [*image_widgets, *ban_widgets]

    # Creating float text widgets for entering weights for images
    image_weight_widgets = {widget: widgets.FloatText(value=1.0, description=widget.description + " W:", layout=widgets.Layout(width='130px')) for widget in image_widgets}

    result_output = widgets.Output()

    # Generate button
    generate_button = widgets.Button(description='Generate')
    generate_button.on_click(generat_prompt_from_input)

    # Create a list of HBoxes to organize the widgets
    dropdown_boxes_list = [
        widgets.HBox([select_widgets_dict[key], widget, image_weight_widgets.get(widget, None)])
        if "image" in widget.description
        else widgets.HBox([select_widgets_dict[key], widget])
        for key, widget in zip(select_widgets_dict.keys(), image_n_ban_widgets)
    ]

    # Display the widgets
    display(*dropdown_boxes_list, generate_button, result_output)
```
